```python
import railfence
```

```python
thequote = "People say nothing is impossible, but I do nothing every day - A. A. Milne"
unquote = railfence.encrypt(thequote, 4, 0)
print(unquote)
```

```python
quote = railfence.decrypt(unquote, 4, 0)
print(quote)
```
